8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
# Coordenadas de Madrid (latitud y longitud)
madrid_lat = 40.416775
madrid_lon = -3.703790

# Calcular la distancia entre cada municipio y Madrid
def haversine(lat1, lon1, lat2, lon2):
    r = 6371  # Radio de la Tierra en kilómetros
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2) ** 2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = r * c
    return distance

dataset["Distancia_a_Madrid"] = haversine(dataset["Y"], dataset["X"], madrid_lat, madrid_lon)

# Excluir Madrid de los resultados
df_sin_madrid = dataset[dataset["Municipio"] != "Madrid"]

# Encontrar el municipio más cercano y más lejano a Madrid
municipio_mas_cercano = df_sin_madrid[df_sin_madrid["Distancia_a_Madrid"] == df_sin_madrid["Distancia_a_Madrid"].min()]["Municipio"].values[0]
municipio_mas_lejano = df_sin_madrid[df_sin_madrid["Distancia_a_Madrid"] == df_sin_madrid["Distancia_a_Madrid"].max()]["Municipio"].values[0]

print(f"El municipio más cercano a Madrid (excluyendo Madrid) es: {municipio_mas_cercano}")
print(f"El municipio más lejano a Madrid (excluyendo Madrid) es: {municipio_mas_lejano}")
```
